Подключим библиотеки, которые будут в дальнейшем использоваться в лабораторной работе:

```python
import numpy as np
import math as m
import pandas as pd #оформление таблицы
from sklearn.datasets import * #получение датасета
import seaborn as sns #визуализация данных
sns.set(style="ticks")
```

Подключим датасет указанный в задании:

```python
boston = load_boston()
data1 = pd.DataFrame(data= np.c_[boston['data']],
                     columns= boston['feature_names'])
data2 = pd.DataFrame(data= np.c_[boston['target']],
                    columns= ['target'])
data = pd.concat([data1, data2], axis=1)
```

Создаем таблицы для основного набора данных и для целевого признака, после чего обхединяем их в одну. Теперь можно работать уже с таблицей.

```python
data
```

Оценим количество пропусков в таблице:

```python
data.isnull().sum()
```

Видим, что в данном датасете нет пропусков, поэтому можно сразу переходить к корреляционному анализу. Для построения таблицы коэффициент корреляции Пирсона. 

```python
corr_data = data.corr(method='pearson')
corr_data
```

Вычислим количество признаков сильно коррелирующих (от 0,7 до 1), средне коррелирующих (от 0,5 до 0,7) и слабо коррелирующих (от 0 до 0,5) с целевым признаком. 
Кроме того определим, какие признаки сильно коррелируют между собой (от 0,7 до 1).

```python
good_cor = []
norm_cor = []
bad_corr = []
corr_nontarg = [] 
cols = []

for col in corr_data:
    cols.append(col)
print(cols)
    
for i in range(0, 14):
    for j in range(0, i):
        cor = m.fabs(corr_data[cols[i]][cols[j]])
        
        if(cor > 0.7):
            corr_nontarg.append("{0} & {1} ({2})".format(cols[i], cols[j], cor))
```
